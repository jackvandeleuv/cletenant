```python
def summarize_with_count(df, keys, values, count_name=None):
    agg = df.groupby(keys)[values].sum().reset_index()
    if not count_name:
        return agg
    
    agg_count = (df
        .groupby(keys)
        .size()
        .reset_index()
        .rename(columns={0: count_name})
    )
    return agg.merge(agg_count, on=keys)

# agg_owners = summarize_with_count(par, ['owner_id'], agg_cols, 'parcels_owned')


# Owners by neighborhood.
owners_by_neighborhood = summarize_with_count(par, ['activerentalregistrationflag', 'neighborhood', 'owner_id'], agg_cols, 'parcels_owned')
owners_by_neighborhood_count = owners_by_neighborhood.groupby(['activerentalregistrationflag', 'neighborhood']).size().reset_index().rename(columns={0: 'owners'})
```
